```python
#import pandas 
import pandas as pd 

#insert dataset
churn = pd.read_csv("churn-bigml-20.csv")

# Set state and city as index modifying the DataFrame
churn.set_index("State", inplace=True)

# Print churn
print(churn.head())
```

```python
#import pandas 
import pandas as pd 

#insert dataset
churn = pd.read_csv("churn-bigml-20.csv")

# Melt publisher column using title and authors as identifiers
churn_melted = churn.melt(id_vars=["State", "Area code"], 
                                     value_vars=["Total day minutes","Total day charge"],var_name="activity", value_name="amount")

# Print publisher_melted
print(churn_melted.head(10))
```

```python
#import pandas 
import pandas as pd 

#insert dataset
churn = pd.read_csv("churn-bigml-20.csv")

# Stack the resulting DataFrame using the first column level
churn_final = churn.stack(level=0)

# Print churn_time
print(churn_final.head(20))
```
