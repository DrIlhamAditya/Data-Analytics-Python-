```python
# dealing with missing data
import pandas as pd
import seaborn as sns

flight_data = pd.read_csv("planes.csv")
# Find the five percent threshold
threshold = len(flight_data) * 0.05
# Create a filter
cols_to_drop = flight_data.columns[flight_data.isna().sum() <= threshold]
# Drop missing values for columns below the threshold
flight_data.dropna(subset=cols_to_drop, inplace=True)
print(flight_data.isna().sum())

# Calculate median plane ticket prices by Airline
airline_prices = flight_data.groupby("Airline")["Price"].median()
# Convert to a dictionary
prices_dict = airline_prices.to_dict()
# Map the dictionary to missing values of Price by Airline
flight_data["Price"] = flight_data["Price"].fillna(flight_data["Airline"].map(prices_dict))
# Check for missing values
print(flight_data.isna().sum())
```

```python
#adding descriptive statistics
import pandas as pd
import seaborn as sns

flight_data = pd.read_csv("planes.csv")
# Price standard deviation by Airline
flight_data["airline_price_st_dev"] = flight_data.groupby("Airline")["Price"].transform(lambda x: x.std())
print(flight_data[["Airline", "airline_price_st_dev"]].value_counts())
```

```python
#adding descriptive statistics
import pandas as pd
import seaborn as sns

flight_data = pd.read_csv("planes.csv")
# Mean Price by Destination
flight_data["price_destination_mean"] = flight_data.groupby("Destination")["Price"].transform(lambda x: x.mean())
print(flight_data[["Destination","price_destination_mean"]].value_counts())
```

```python
# dealing with outliers
import pandas as pd
import seaborn as sns

flight_data = pd.read_csv("planes.csv")# Find the 75th and 25th percentiles
price_seventy_fifth = flight_data["Price"].quantile(0.75)
price_twenty_fifth = flight_data["Price"].quantile(0.25)

# Calculate iqr
prices_iqr = price_seventy_fifth - price_twenty_fifth

# Calculate the thresholds
upper = price_seventy_fifth + (1.5 * prices_iqr)
lower = price_twenty_fifth - (1.5 * prices_iqr)

# Subset the data
flight_data = flight_data[(flight_data["Price"] > lower) & (flight_data["Price"] < upper)]

print(flight_data["Price"].describe())
```
