```python
#import pandas 
import pandas as pd 

#insert dataset
fifa_data = pd.read_csv("players_20.csv")

# Remove duplicates to avoid pivot error
fifa_data = fifa_data.drop_duplicates(subset=["short_name", "skill_moves"])

# Pivot fifa_players to get overall scores indexed by name and identified by movement
fifa_overall = fifa_data.pivot(index="short_name", columns="skill_moves", values="overall")

# Print fifa_overall
print(fifa_overall.head(10))
```

```python
#import pandas 
import pandas as pd 

#insert dataset
fifa_data = pd.read_csv("players_20.csv")

# Melt books_gothic using the title, authors, and publisher columns as identifier
fifa_melted = fifa_data.melt(id_vars=["club","nationality","player_positions"])

# Print gothic_melted_new
print(fifa_melted.head(10))
```

```python
#import pandas 
import pandas as pd 

#insert dataset
fifa_data = pd.read_csv("players_20.csv")

# Set state and city as index modifying the DataFrame
fifa_data.set_index(["club", "short_name"], inplace=True)

# Switch the first and third row index levels in churn
fifa_swap = fifa_data.swaplevel(0, 1)

# Print churn
print(fifa_swap)
```
