```python
#unboxing for mean, median, and mode of the quantity column 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

print(dataset.loc[:, 'quantity'].mean())
print(dataset.loc[:, 'quantity'].median())
print(dataset.loc[:, 'quantity'].mode())
```

```python
#unboxing for mean, median, and mode of the item_price column 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

print(dataset.loc[:, 'item_price'].mean())
print(dataset.loc[:, 'item_price'].median())
print(dataset.loc[:, 'item_price'].mode())
```

```python
#standar deviation and variance of quantity column 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

print(dataset.loc[:, 'quantity'].std())
print(dataset.loc[:, 'quantity'].var())
```

```python
#standar deviation and variance of item_price column 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

print(dataset.loc[:, 'item_price'].std())
print(dataset.loc[:, 'item_price'].var())
```

```python
#descending sort of quantity data 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

dataset.sort_values(by = 'quantity', ascending = False)
```

```python
#ascending sort of quantity data 
import pandas as pd 

dataset = pd.read_csv('retail_raw.csv')

dataset.sort_values(by = 'quantity') #ascending = True 
```

```python
#identify the outliers 1
import pandas as pd 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('retail_raw.csv')

dataset[['quantity']].hist(figsize=(3, 4), bins=7, xlabelsize=6, ylabelsize=6)
dataset[['item_price']].hist(figsize=(3, 4), bins=7, xlabelsize=6, ylabelsize=6)
plt.show()
```

```python
#how to detect outliers using boxplot 1 
import pandas as pd 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('retail_raw.csv')

dataset[['quantity']].plot(kind='box')
plt.title('Box Plot of Quantity', size=10)
plt.show()
```

```python
#how to detect outliers using boxplot 2
import pandas as pd 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('retail_raw.csv')

dataset[['item_price']].plot(kind='box')
plt.title('Box Plot of Price Items', size=10)
plt.show()
```
